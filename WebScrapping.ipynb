```python
from typing import Text
from pandas.io.formats.format import TextAdjustment

#Obtener datos búqueda en google

def obtener_descripcion(enlaces):
  url = enlaces
  descripciones_encontradas = []
  descripciones_encontradas.clear()

  try:
    respuesta1 = requests.get(url, headers = headers, cert=("/usr/local/lib/python3.7/dist-packages/certifi/cacert.pem"))
    statusCode = respuesta.status_code
    if statusCode == 200:
      soup_response1 = soup(respuesta1.text, 'html.parser')
      texto = soup_response1.find_all("p")
      for resultado1 in texto:
        if resultado1.find('p') is not None:
            descripcion = resultado1.find('p').get_text()
            descripciones_encontradas.append(descripcion)
        else:
          descripciones_encontradas.append('')
    else: descripciones_encontradas.append('')
  except requests.exceptions.ConnectionError as e:
     descripciones_encontradas.append('Visita página web')
  
  return descripciones_encontradas

def obtener_resultados(termino_busqueda, numero_resultados, codigo_lenguaje):
  url_google = 'https://www.google.com/search?q={}&num={}&h1={}'.format(termino_busqueda, numero_resultados, codigo_lenguaje)
  respuesta = requests.get(url_google, headers = headers)
  respuesta.raise_for_status()
  return termino_busqueda, respuesta.text

def procesar_resultados(html, palabra):
  soup_response = soup(html, 'html.parser')
  bloque = soup_response.find_all("div", class_="g")
  
  for resultado in bloque:
    titulo = resultado.find('h3').get_text()
    enlaces = resultado.find('a',href=True).get('href')
    if resultado.find('div', {"class": "VwiC3b yXK7lf MUxGbd yDYNvb lyLwlc lEBKkf"}) is not None:
        descripciones = resultado.find('div', {"class": "VwiC3b yXK7lf MUxGbd yDYNvb lyLwlc lEBKkf"}).get_text()
    else:
      descripciones = 'Visita página web'
    global df
    df = df.append({'palabra clave' : palabra, 
                    'titulo' : titulo, 
                    'enlace' : enlaces, 
                    'descripcion' : descripciones},
                    ignore_index = True)   
  return df


def scrap(termino_busqueda, numero_resultados, codigo_lenguaje):
  palabra, html = obtener_resultados(termino_busqueda, numero_resultados, codigo_lenguaje)
  resultados = procesar_resultados(html, palabra)
  return resultados

if __name__ == '__main__':
  convocatorias = ['convocatoria social', 'proyecto social', 'cooperacion social', 'apoyo economico', 'apoyo social', 'ayuda social', 'alianza social', 
                   'concurso social', 'innovacion social', 'programa social']
  clave = ['salud', 'educacion', 'niñez', 'alimentacion', 'niños', 'aprendizaje', 'adolescentes', 'ambiental', 'seguridad', 'comunitario', 'comunitaria', 
           'niñas', 'ciencia', 'tecnologia', 'naciones unidas', 'becas', 'jóvenes', 'alfabetización', 'sexualidad', 'salubridad', 'ONG', 'vejez', 
           'hambre', 'desastre', 'migrantes', 'discapacidad', 'estudiantes', 'vulnerabilidad']
  palabra = []

  for i in convocatorias:
    for j in clave:
      concat = str(i+" + "+j)
      palabra.append(concat)

  resultados = pd.DataFrame(columns = ['palabra clave', 'titulo', 'enlace', 'descripcion'])  
  
  for i in palabra:
    resultados = pd.concat([resultados, scrap(i, 5, "es")])
```
